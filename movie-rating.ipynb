```python
def process_movie_ratings(ratings_df, movies_df):
    # Merge the data
    ratings_with_movie = pd.merge(ratings_df, movies_df, on='movieId', how='left')

    # Filter out movies with less than 1000 ratings
    rating_counts = ratings_with_movie['movieId'].value_counts()
    movies_to_remove = rating_counts[rating_counts < 1000].index.tolist()
    filtered_ratings = ratings_with_movie[~ratings_with_movie['movieId'].isin(movies_to_remove)]

    # Create pivot table
    pivot_table = filtered_ratings.pivot_table(index='userId', columns='title', values='rating')

    return pivot_table
```
